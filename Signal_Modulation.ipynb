# **Plotten Sie das Spektrum eines amplitudenmodulierten Signals mit Datenrate D=1 kBit/s und Frequenz des Trägersignals ftr = 10 kHz.**

Lösung:

**Länge eines Bits: Tb = 1/D =>    Tb=1ms**

**Hauptperiode: Tp=2*Tb  =>   Tp=2ms**

**Hauptrequenz: f0=1/Tp  =>   f0=500Hz**

**Frequenz des Trägersignals: ftr = 10kHz = 10 000 Hz**

 Import libraries:

```python
from scipy import signal
import matplotlib.pyplot as plot
import numpy as np
from numpy.fft import fft, fftfreq
```

# Darstellung des rechteckigen Signals (Binarysignal)

```python
#number of samples for the given interval
ns = 1000 

#end of the interval
ei = 0.004 

#sample rate
sr = ns/ei 

#frequency binary signal
fbin = 500 

time = np.linspace(0, ei, ns)
BinSig = (signal.square(2 * np.pi * fbin * time) + 1.0)/2
plot.plot(time,BinSig)

plot.xlabel('Time')
plot.ylabel('Amplitude')
plot.title('Das recheckige Signal')
```

# Darstellung der cos-Funktion (Trägersignal)

```python
#frequency of the cos signal 
CosFreq = 10000

CosSig = np.cos(2 * np.pi * time * CosFreq)
plot.plot(time, CosSig)

plot.xlabel('Time')
plot.ylabel('Amplitude')
plot.title('Die Cos - Funktion (Trägersignal)')
```

# Darstellung des amplitudenmodulierten Signals

```python
ModSig = BinSig * CosSig
plot.plot(time, ModSig)

plot.xlabel('Time')
plot.ylabel('Amplitude')
plot.title('Das amplitudenmodulierte Signal')
```

# Darstellung des Spektrums des rechteckigen Signals

```python
#Spektrum des rechteckigen Signals

yr = fft(BinSig)
xr = fftfreq(len(BinSig))

plot.xlabel('Frequenz in Hz')
plot.ylabel('Amplitude')
plot.title('Recheckiges Signal')

plot.stem(xr*sr, np.abs(yr),markerfmt=" ", use_line_collection=True)     
plot.xlim(-sr/10,sr/10)
plot.show()
```

# Darstellung des Spektrums des cos - Signals

```python
#Spektrum des cos-Signals

Nc = len(fft(CosSig))          
yc = fft(CosSig);          
xc = fftfreq(len(CosSig)) 

plot.xlabel('Frequenz in Hz')
plot.ylabel('Amplitude')
plot.title('Cos-Signal')

plot.stem(xc*sr,  np.abs(yc),markerfmt=" ", use_line_collection=True)
plot.xlim(-sr/10,sr/10)
plot.show()
```

# Darstellung des Spektrums des amplitudenmodulierten Signals



```python
#Spectrum of the amplitude molilated signal

ym = fft(ModSig); 
xm = fftfreq(len(ModSig))

plot.xlabel('Frequenz in Hz')
plot.ylabel('Amplitude')
plot.title('Amplitudenmoduliertes Signal')

plot.stem(xm*sr, np.abs(ym),markerfmt=" ", use_line_collection=True); 
plot.xlim(10000,11000)
plot.show()
```
